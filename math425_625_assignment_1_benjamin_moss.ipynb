**Question 1:** Suppose that $fl(y)$ is the $k$-digit rounding (i.e., not simple chopping) approximation to $y$. Show that

$$
\left| \dfrac{y - fl(y)}{y}\right| \leq 0.5 \times 10^{-k + 1}
$$

**Question 1:** Suppose that $fl(y)$ is the $k$-digit rounding (i.e., not simple chopping) approximation to $y$. We will show that 
$$\bigg\lvert\dfrac{y-fl(y)}{y}\bigg\rvert\leq0.5\times10^{-k+1}.$$

**Proof:** We must consider two cases, one in which our figure $d_{k+1}<5$ and the other in which $d_{k+1}\geq 5.$ From our textbook, we know that this corresponds to when $fl(y)=0.d_{1}d_{2}\cdots\,d_{k}\times10^{n}$ and when $fl(y)=0.d_{1}d_{2}\cdots\,d_{k}\times10^{n}+10^{n-k}$, respectively. Then we proceed with our first case. Let $d_{k+1}<5$, then we have 
$$
\bigg\lvert\dfrac{y-fl(y)}{y}\bigg\rvert=\bigg\lvert\dfrac{\{0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\times10^{n}\}-\{0.d_{1}d_{2}\cdots\,d_{k}\times10^{n}\}}{\{0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\times10^{n}\}}\bigg\rvert \\
=\bigg\lvert\dfrac{\{0.\overbrace{000\cdots0}^{\text{$k$ many 0's}}d_{k+1}d_{k+2}\cdots\times10^{n}\}}{\{0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\times10^{n}\}}\bigg\rvert
$$
Note that subtraction of the terms in the numerator yields a number containing with $d_{1}\cdots\,d_{k}$ many zeros. We proceed and we have 
$$
\bigg\lvert\dfrac{y-fl(y)}{y}\bigg\rvert=\bigg\lvert\dfrac{\{0.d_{k+1}d_{k+2}\cdots\times10^{n-k}\}}{\{0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\times10^{n}\}}\bigg\rvert
$$
where we have rewritten the numerator in normalized decimal form by varying the exponent of base $10$. Then we will rewrite the terms by separating the fractional and exponential parts, giving us 
$$
\bigg\lvert\dfrac{y-fl(y)}{y}\bigg\rvert=\bigg\lvert\dfrac{\{0.d_{k+1}d_{k+2}\cdots\}\times10^{n-k}}{\{0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\}\times10^{n}}\bigg\rvert
\\
\text{now we will rewrite the exponent from the denominator in the numerator}
\\
=\bigg\lvert\dfrac{\{0.d_{k+1}d_{k+2}\cdots\}\times10^{n-k-n}}{\{0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\}}\bigg\rvert,
$$
we simplify and we have 
$$
\bigg\lvert\dfrac{y-fl(y)}{y}\bigg\rvert=\bigg\lvert\dfrac{\{0.d_{k+1}d_{k+2}\cdots\}}{\{0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\}}\bigg\rvert\times10^{-k}.
$$
Now if $d_{k+1}<5$ then this implies that $\{0.d_{k+1}d_{k+2}\}<0.5$ and we have 
$$
\bigg\lvert\dfrac{y-fl(y)}{y}\bigg\rvert\leq\bigg\lvert\dfrac{0.5}{\{0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\}}\bigg\rvert\times10^{-k},
$$
while if $d_{1}\geq1$ then $\{0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\}\geq0.1$ which yields 
$$
\bigg\lvert\dfrac{y-fl(y)}{y}\bigg\rvert\leq\bigg\lvert\dfrac{0.5}{0.1}\bigg\rvert\times10^{-k}\\
\leq0.5\times10^{-k+1}.
$$
Then we have if $d_{k+1}<5\Rightarrow \bigg\lvert\dfrac{y-fl(y)}{y}\bigg\rvert\leq0.5\times10^{-k+1}.$



Now suppose $d_{k+1}\geq5.$ We have 
$$
\bigg\lvert\dfrac{y-fl(y)}{y}\bigg\rvert=\bigg\lvert\dfrac{\{0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\times10^{n}\}-\{0.d_{1}d_{2}\cdots\,d_{k}\times10^{n}+10^{n-k}\}}{\{0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\times10^{n}\}}\bigg\rvert \\
=\bigg\lvert\dfrac{(0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\times10^{n})-(0.d_{1}d_{2}\cdots\,d_{k}+10^{-k})10^{n}}{(0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\times10^{n})}\bigg\rvert\\
=\bigg\lvert\dfrac{(0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\times10^{n})-(0.\delta_{1}\delta_{2}\cdots\,\delta_{k}\times10^{n})}{(0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\times10^{n})}\bigg\rvert,
$$
where $\delta_{i}=d_{i}$ for $i=1,2,\cdots,k$ since $d_{k+1}\geq5.$ Then we have
$$
\bigg\lvert\dfrac{y-fl(y)}{y}\bigg\rvert=\bigg\lvert\dfrac{[(0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2})-(0.\delta_{1}\delta_{2}\cdots\,\delta_{k})]10^{n}}{(0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots\times10^{n})}\bigg\rvert\\
=\bigg\lvert\dfrac{(0.\overbrace{00\cdots0}^{\text{$k-1$ many 0's}}d_{k}d_{k+1}d_{k+2}\cdots)-(0.\overbrace{00\cdots0}^{\text{$k-1$ many 0's}}\delta_{k})}{(0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots)}\bigg\rvert\\
=\bigg\lvert\dfrac{0.\overbrace{000\cdots0}^{\text{$k-1$ many 0's}}05}{(0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots)}\bigg\rvert,
$$
the above is due to the fact that $0.000\cdots0\delta_{k}>0.000\cdots\,d_{k}d_{k+1}d_{k+2}\cdots$ and $\delta_{k}=d_{k}+1$ for $d\geq5.$ From here we rewrite the numerator and we have 
$$
\bigg\lvert\dfrac{y-fl(y)}{y}\bigg\rvert=\bigg\lvert\dfrac{0.5\times10^{-k}}{(0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots)}\bigg\rvert\\
\leq\bigg\lvert\dfrac{0.5\times10^{-k}}{0.1}\bigg\rvert
\text{since $d_{1}\geq1\Rightarrow(0.d_{1}d_{2}\cdots\,d_{k}d_{k+1}d_{k+2}\cdots)\geq0.1$, we proceed and simplify and we have}\\
\bigg\lvert\dfrac{y-fl(y)}{y}\bigg\rvert\leq\bigg\lvert\dfrac{0.5\times10^{-k}}{0.1}\bigg\rvert\\
\leq0.5\times10^{-k+1}.
$$
Thus, if $d_{k+1}\geq5\Rightarrow\bigg\lvert\dfrac{y-fl(y)}{y}\bigg\rvert\leq0.5\times10^{-k+1}.$ Therefore, from both of these cases, the result follows. $\blacksquare$
